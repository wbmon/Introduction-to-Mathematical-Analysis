## 1. Найти интервалы возрастания и убывания функций:

### 1) $ f(x) = x + e^{-x} $

Найдем производную функции.

$$ f'(x) = (x + e^{-x})' = 1 - e^{-x} $$

Определим, где производная положительна и где отрицательна.

- $ f'(x) > 0 $ при $ 1 - e^{-x} > 0 $
  $$ e^{-x} < 1 $$
  $$ -x < 0 $$
  $$ x > 0 $$

- $ f'(x) < 0 $ при $ 1 - e^{-x} < 0 $
  $$ e^{-x} > 1 $$
  $$ -x > 0 $$
  $$ x < 0 $$

Значит:

- функция возрастает на интервале $ (0, +\infty) $.
- функция убывает на интервале $ (-\infty, 0) $.

### 2) $ f(x) = x \ln x $

Найдем производную функции.

$$ f'(x) = (x \ln x)' = \ln x + 1 $$

Определим интервалы знакопостоянства производной.

- $ f'(x) > 0 $ при $ \ln x + 1 > 0 $
  $$ \ln x > -1 $$
  $$ x > e^{-1} $$
  $$ x > \frac{1}{e} $$

- $ f'(x) < 0 $ при $ \ln x + 1 < 0 $
  $$ \ln x < -1 $$
  $$ x < e^{-1} $$
  $$ x < \frac{1}{e} $$

Получаем:

- функция возрастает на интервале $ \left( \frac{1}{e}, +\infty \right) $.
- функция убывает на интервале $ (0, \frac{1}{e}) $.

### 3) $ y = \frac{1}{1 - x^2} $

Найдем производную функции.

$$ y' = \left( \frac{1}{1 - x^2} \right)' = \left( (1 - x^2)^{-1} \right)' = - (1 - x^2)^{-2} \cdot (-2x) = \frac{2x}{(1 - x^2)^2} $$

Определим, где производная положительна и где отрицательна.

- $ y' > 0 $ при $ \frac{2x}{(1 - x^2)^2} > 0 $
  $$ 2x > 0 $$
  $$ x > 0 $$

- $ y' < 0 $ при $ \frac{2x}{(1 - x^2)^2} < 0 $
  $$ 2x < 0 $$
  $$ x < 0 $$

Значит:

- функция возрастает на интервале $ (0, 1) \cup (1, +\infty) $.
- функция убывает на интервале $ (-\infty, -1) \cup (-1, 0) $.


#### **Ответ:**

1) $ f(x) = x + e^{-x} $:
   - Возрастает на $ (0, +\infty) $
   - Убывает на $ (-\infty, 0) $

2) $ f(x) = x \ln x $:
   - Возрастает на $ \left( \frac{1}{e}, +\infty \right) $
   - Убывает на $ (0, \frac{1}{e}) $

3) $ y = \frac{1}{1 - x^2} $:
   - Возрастает на $ (0, 1) \cup (1, +\infty) $
   - Убывает на $ (-\infty, -1) \cup (-1, 0) $





## 2. Найти экстремумы функций:

$$ f'(x) = (x^3 - 3x + 1)' = 3x^2 - 3 $$

Приравняем производную к нулю и найдем критические точки.

$$ 3x^2 - 3 = 0 $$
$$ 3(x^2 - 1) = 0 $$
$$ x^2 - 1 = 0 $$
$$ x^2 = 1 $$
$$ x = \pm 1 $$

Определим знаки производной в интервалах между критическими точками.

 Для $ x < -1 $: $ f'(x) = 3x^2 - 3 > 0 $  
 Для $ -1 < x < 1 $: $ f'(x) = 3x^2 - 3 < 0 $  
 Для $ x > 1 $: $ f'(x) = 3x^2 - 3 > 0 $

Определим экстремумы.

 При $ x = -1 $: $ f(-1) = (-1)^3 - 3(-1) + 1 = -1 + 3 + 1 = 3 $ (максимум)  
 При $ x = 1 $: $ f(1) = 1^3 - 3(1) + 1 = 1 - 3 + 1 = -1 $ (минимум)

$$ x = 2 $$

Определим экстремум.

 При $ x = 2 $: $ y(2) = e^{2^2 - 4 \cdot 2 + 5} = e^{4 - 8 + 5} = e^1 = e $ (минимум)

## 3. Найти интервалы выпуклости и точки перегиба функций:

### 1) $ f(x) = e^{-x^2} $

Найдем первую производную функции.

$$ f'(x) = (e^{-x^2})' = e^{-x^2} \cdot (-2x) = -2x e^{-x^2} $$

Найдем вторую производную функции.

$$ f''(x) = (-2x e^{-x^2})' = -2 e^{-x^2} + (-2x) \cdot (-2x e^{-x^2}) = -2 e^{-x^2} + 4x^2 e^{-x^2} $$
$$ f''(x) = e^{-x^2} (4x^2 - 2) $$

Приравняем вторую производную к нулю и найдем критические точки.

$$ e^{-x^2} (4x^2 - 2) = 0 $$
$$ 4x^2 - 2 = 0 $$
$$ 4x^2 = 2 $$
$$ x^2 = \frac{1}{2} $$
$$ x = \pm \frac{1}{\sqrt{2}} $$

Определим знаки второй производной в интервалах между критическими точками.

 Для $ x < -\frac{1}{\sqrt{2}} $: $ f''(x) = e^{-x^2} (4x^2 - 2) < 0 $  
 Для $ -\frac{1}{\sqrt{2}} < x < \frac{1}{\sqrt{2}} $: $ f''(x) = e^{-x^2} (4x^2 - 2) > 0 $  
 Для $ x > \frac{1}{\sqrt{2}} $: $ f''(x) = e^{-x^2} (4x^2 - 2) < 0 $  

Определим интервалы выпуклости и точки перегиба.

 Функция выпукла вверх на интервалах $ (-\infty, -\frac{1}{\sqrt{2}}) $ и $ (\frac{1}{\sqrt{2}}, \infty) $.  
 Функция выпукла вниз на интервале $ (-\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}) $.  
 Точки перегиба: $ x = \pm \frac{1}{\sqrt{2}} $.  

### 2) $ y = \cos x $

$$ y' = (\cos x)' = -\sin x $$

Найдем вторую производную функции.

$$ y'' = (-\sin x)' = -\cos x $$

Приравняем вторую производную к нулю и найдем критические точки.

$$ -\cos x = 0 $$
$$ \cos x = 0 $$
$$ x = \frac{\pi}{2} + k\pi \quad (k \in \mathbb{Z}) $$

Определим знаки второй производной в интервалах между критическими точками.

 Для $ 0 < x < \frac{\pi}{2} $: $ y'' = -\cos x < 0 $  
 Для $ \frac{\pi}{2} < x < \pi $: $ y'' = -\cos x > 0 $  

Определим интервалы выпуклости и точки перегиба.

 Функция выпукла вверх на интервалах $ (2k\pi - \frac{\pi}{2}, 2k\pi + \frac{\pi}{2}) $.  
 Функция выпукла вниз на интервалах $ (2k\pi + \frac{\pi}{2}, 2k\pi + \frac{3\pi}{2}) $.  
 Точки перегиба: $ x = \frac{\pi}{2} + k\pi \quad (k \in \mathbb{Z}) $.

## 4. Найти асимптоты графиков функций:

### 1) $ y = \frac{3x}{x+2} $

Найдем вертикальные асимптоты.

Вертикальные асимптоты могут возникнуть там, где знаменатель функции обращается в ноль.

$$ x + 2 = 0 $$
$$ x = -2 $$

Таким образом, вертикальная асимптота: $ x = -2 $.

Найдем горизонтальные асимптоты.

Для нахождения горизонтальных асимптот, рассмотрим поведение функции при $ x \to \pm \infty $.

$$ y = \frac{3x}{x+2} = \frac{3x}{x(1 + \frac{2}{x})} = \frac{3}{1 + \frac{2}{x}} $$

При $ x \to \pm \infty $:

$$ y \to \frac{3}{1 + 0} = 3 $$

Таким образом, горизонтальная асимптота: $ y = 3 $.

### 2) $ y = e^{-\frac{1}{x}} $

Найдем вертикальные асимптоты.

Вертикальные асимптоты могут возникнуть там, где выражение под экспонентой обращается в бесконечность.

$$ -\frac{1}{x} \to \infty \quad \text{при} \quad x \to 0^+ $$
$$ -\frac{1}{x} \to -\infty \quad \text{при} \quad x \to 0^- $$

Таким образом, вертикальная асимптота: $ x = 0 $.

Найдем горизонтальные асимптоты.

Для нахождения горизонтальных асимптот, рассмотрим поведение функции при $ x \to \pm \infty $.

При $ x \to \infty $:

$$ e^{-\frac{1}{x}} \to e^0 = 1 $$

При $ x \to -\infty $:

$$ e^{-\frac{1}{x}} \to e^0 = 1 $$

Таким образом, горизонтальная асимптота: $ y = 1 $.

### Ответ:

1) $ y = \frac{3x}{x+2} $:  
    Вертикальная асимптота: $ x = -2 $  
    Горизонтальная асимптота: $ y = 3 $  

2) $ y = e^{-\frac{1}{x}} $:  
    Вертикальная асимптота: $ x = 0 $  
    Горизонтальная асимптота: $ y = 1 $  

## 5. Провести полное исследование и построить графики функций:

### 1) $ y = \ln(1 - x^2) $

**Область определения:**
$$ 1 - x^2 > 0 $$
$$ -1 < x < 1 $$

**Асимптоты:**  
 Вертикальные асимптоты: $ x = -1 $ и $ x = 1 $ (там, где аргумент логарифма обращается в ноль).  
 Горизонтальные асимптоты: нет, так как $ \ln(1 - x^2) \to -\infty $ при $ x \to \pm 1 $.  

**Точки пересечения с осями:**
- С осью $ x $ (при $ y = 0 $):
$$ \ln(1 - x^2) = 0 $$
$$ 1 - x^2 = 1 $$
$$ x^2 = 0 $$
$$ x = 0 $$
- С осью $ y $ (при $ x = 0 $):
$$ y = \ln(1 - 0^2) = \ln(1) = 0 $$

**Производная и экстремумы:**
$$ y' = \left( \ln(1 - x^2) \right)' = \frac{-2x}{1 - x^2} $$
- Критические точки: $ y' = 0 $
$$ -2x = 0 $$
$$ x = 0 $$
- Исследуем знак производной:
  - При $ x < 0 $: $ y' > 0 $ (функция возрастает)
  - При $ x > 0 $: $ y' < 0 $ (функция убывает)
- Экстремум: $ x = 0 $ — максимум, $ y(0) = 0 $.

**Вторая производная и точки перегиба:**
$$ y'' = \left( \frac{-2x}{1 - x^2} \right)' = \frac{-2(1 - x^2) + 2x \cdot 2x}{(1 - x^2)^2} = \frac{-2 + 6x^2}{(1 - x^2)^2} $$
- Точки перегиба: $ y'' = 0 $
$$ -2 + 6x^2 = 0 $$
$$ 6x^2 = 2 $$
$$ x^2 = \frac{1}{3} $$
$$ x = \pm \frac{1}{\sqrt{3}} $$

**График:**
График функции $ y = \ln(1 - x^2) $ симметричен относительно оси $ y $, возрастает при $ x < 0 $ и убывает при $ x > 0 $. Он имеет максимум в точке $ (0, 0) $ и точки перегиба в $ x = \pm \frac{1}{\sqrt{3}} $.

### 2) $ y = \frac{x^2}{1 - x^2} $

**Область определения:**
$ 1 - x^2 \neq 0 $
$ x \neq \pm 1 $

**Асимптоты:**
- Вертикальные асимптоты: $ x = -1 $ и $ x = 1 $.
- Горизонтальные асимптоты: $ y \to -1 $ при $ x \to \pm \infty $.

**Точки пересечения с осями:**
- С осью $ x $ (при $ y = 0 $):
$$ \frac{x^2}{1 - x^2} = 0 $$
$$ x^2 = 0 $$
$$ x = 0 $$ 
- С осью $ y $ (при $ x = 0 $):  
$$ y = \frac{0^2}{1 - 0^2} = 0 $$

**Производная и экстремумы:**
$$ y' = \frac{(1 - x^2) \cdot 2x - x^2 \cdot (-2x)}{(1 - x^2)^2} = \frac{2x - 2x^3 + 2x^3}{(1 - x^2)^2} = \frac{2x}{(1 - x^2)^2} $$
- Критические точки: $ y' = 0 $  
$$ 2x = 0 $$
$$ x = 0 $$
- Исследуем знак производной:
  - При $ x < 0 $: $ y' < 0 $ (функция убывает)
  - При $ x > 0 $: $ y' > 0 $ (функция возрастает)
- Экстремум: $ x = 0 $ — минимум, $ y(0) = 0 $.

**Вторая производная и точки перегиба:**
$$ y'' = \left( \frac{2x}{(1 - x^2)^2} \right)' = \frac{2(1 - x^2)^2 - 2x \cdot 2(1 - x^2)(-2x)}{(1 - x^2)^4} = \frac{2(1 - x^2) + 8x^2}{(1 - x^2)^3} = \frac{2 + 6x^2}{(1 - x^2)^3} $$
- Точки перегиба: 
$$ y'' = 0 $$
$$ 2 + 6x^2 = 0 $$
$$ 6x^2 = -2 $$
Нет реальных точек перегиба.

**График:**
График функции $ y = \frac{x^2}{1 - x^2} $ симметричен относительно оси $ y $, убывает при $ x < 0 $ и возрастает при $ x > 0 $. Он имеет минимум в точке $ (0, 0) $ и асимптоты $ x = \pm 1 $ и $ y = -1 $.



### 3) $ y = x^2 \cdot e^{-x} $

**Область определения:**
Все $ x \in \mathbb{R} $.

**Асимптоты:**
- Вертикальные асимптоты: нет.
- Горизонтальные асимптоты: $ y \to 0 $ при $ x \to \infty $.

**Точки пересечения с осями:**
- С осью $ x $ (при $ y = 0 $):
$$ x^2 \cdot e^{-x} = 0 $$
$$ x^2 = 0 $$
$$ x = 0 $$
- С осью $ y $ (при $ x = 0 $):
$$ y = 0^2 \cdot e^0 = 0 $$

**Производная и экстремумы:**
$$ y' = (x^2 \cdot e^{-x})' = 2x \cdot e^{-x} + x^2 \cdot (-e^{-x}) = e^{-x}(2x - x^2) $$
- Критические точки: $ y' = 0 $  
$$ e^{-x}(2x - x^2) = 0 $$
$$ 2x - x^2 = 0 $$
$$ x(2 - x) = 0 $$
$$ x = 0 \text{ или } x = 2 $$
- Исследуем знак производной:
  - При $ x < 0 $: $ y' < 0 $ (функция убывает)
  - При $ 0 < x < 2 $: $ y' > 0 $ (функция возрастает)
  - При $ x > 2 $: $ y' < 0 $ (функция убывает)
- Экстремумы: $ x = 0 $ — минимум, $ y(0) = 0 $; $ x = 2 $ — максимум, $ y(2) = 4e^{-2} $.

**Вторая производная и точки перегиба:**
$$ y'' = \frac{d}{dx} \left( e^{-x}(2x - x^2) \right) = e^{-x}(2 - 2x) + e^{-x}(-2x + x^2) = e^{-x}(2 - 4x + x^2) $$
- Точки перегиба: $ y'' = 0 $
$$ 2 - 4x + x^2 = 0 $$
$$ x^2 - 4x + 2 = 0 $$
$$ x = 2 \pm \sqrt{2} $$

**График:**
График функции $ y = x^2 \cdot e^{-x} $ убывает при $ x < 0 $, возрастает при $ 0 < x < 2 $, убывает при $ x > 2 $. Он имеет минимум в точке $ (0, 0) $, максимум в точке $ (2, 4e^{-2}) $ и точки перегиба в $ x = 2 \pm \sqrt{2} $.

График функции $ y = x - \ln x $ убывает при $ 0 < x < 1 $, возрастает при $ x > 1 $. Он имеет минимум в точке $ (1, 1) $ и вертикальную асимптоту $ x = 0 $.

```python
import numpy as np
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(12, 12))

# y = ln(1 - x^2)
x1 = np.linspace(-0.99, 0.99, 400)
y1 = np.log(1 - x1**2)

def f1(x):
    return np.log(1 - x**2)


axs[0, 0].plot(x1, y1, label='$y = \\ln(1 - x^2)$')
axs[0, 0].set_title('График функции $y = \\ln(1 - x^2)$')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].axvline(x=-1, color='red', linestyle='dashed', label='Асимптота x = -1')
axs[0, 0].axvline(x=1, color='red', linestyle='dashdot', label='Асимптота x = 1')
axs[0, 0].plot(0, 0, 'go', label='Критическая точка (0, 0)')

roots = [-1/np.sqrt(3), 1/np.sqrt(3)]
for root in roots:
    axs[0, 0].plot(root, f1(root), 'brown', marker='o', label=f'Точка перегиба ({root:.2f}, {f1(root):.2f})')

axs[0, 0].axhline(0, color='black', linewidth=0.5)
axs[0, 0].axvline(0, color='black', linewidth=0.5)
axs[0, 0].legend()

# y = x^2 / (1 - x^2)
x2 = np.linspace(-1.6, 1.6, 400)
y2 = x2**2 / (1 - x2**2)

axs[0, 1].plot(x2, y2, label='$y = \\frac{x^2}{1 - x^2}$')
axs[0, 1].set_title('График функции $y = \\frac{x^2}{1 - x^2}$')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
axs[0, 1].axvline(x=-1, color='red', linestyle='dashed', label='Асимптота x = -1')
axs[0, 1].axvline(x=1, color='red', linestyle='dashdot', label='Асимптота x = 1')
axs[0, 1].axhline(y=-1, color='red', linestyle='dotted', label='Асимптота y = -1')
axs[0, 1].plot(0, 0, 'go', label='Критическая точка (0, 0)')
axs[0, 1].axhline(0, color='black', linewidth=0.5)
axs[0, 1].axvline(0, color='black', linewidth=0.5)
axs[0, 1].legend()

# y = x^2 * e^(-x)
x3 = np.linspace(-1, 6, 400)
y3 = x3**2 * np.exp(-x3)

def f3(x):
    return x**2 * np.exp(-x)

def f3_prime(x):
    return (2*x - x**2) * np.exp(-x)

def f3_double_prime(x):
    return (2 - 4*x + x**2) * np.exp(-x)

axs[1, 0].plot(x3, y3, label='$y = x^2 \\cdot e^{-x}$')
axs[1, 0].set_title('График функции $y = x^2 \\cdot e^{-x}$')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 0].axhline(y=0, color='red', linestyle='dashdot', label='Асимптота y = 0')
axs[1, 0].plot(0, 0, 'go', label='Критическая точка (0, 0)')
axs[1, 0].plot(2, 4*np.exp(-2), 'go', label='Критическая точка (2, 4e^-2)')

roots = fsolve(f3_double_prime, [1, 3])
for root in roots:
    if -1 < root < 6:
        axs[1, 0].plot(root, f3(root), 'brown', marker='o', label=f'Точка перегиба ({root:.2f}, {f3(root):.2f})')

axs[1, 0].axhline(0, color='black', linewidth=0.5)
axs[1, 0].axvline(0, color='black', linewidth=0.5)
axs[1, 0].legend()

# y = x - ln(x)
x4 = np.linspace(0.01, 5, 400)
y4 = x4 - np.log(x4)

axs[1, 1].plot(x4, y4, label='$y = x - \\ln(x)$')
axs[1, 1].set_title('График функции $y = x - \\ln(x)$')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')
axs[1, 1].axvline(x=0, color='red', linestyle='dashdot', label='Асимптота x = 0')
axs[1, 1].plot(1, 1, 'go', label='Критическая точка (1, 1)')
axs[1, 1].axhline(0, color='black', linewidth=0.5)
axs[1, 1].axvline(0, color='black', linewidth=0.5)
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```
