## ***Тема “Предел функции”***

**1. Предложить пример функции, не имеющей предела в нуле и в бесконечностях.**

$$f(x) = \sin\left(\frac{1}{x}\right)$$

Рассмотрим функцию $ f(x) = \sin\left(\frac{1}{x}\right) $. Эта функция не имеет предела ни в нуле, ни в бесконечностях.

**Предел в нуле:**

**Поведение функции при $ x \to 0 $:**
   - При $ x \to 0 $, $ \frac{1}{x} \to \infty $.
   - Функция $ \sin(t) $ является периодической с периодом $ 2\pi $, и при $ t \to \infty $ она не стремится ни к какому пределу, а колеблется между -1 и 1.
   - Следовательно, $ \sin\left(\frac{1}{x}\right) $ не имеет предела при $ x \to 0 $.

**Предел в бесконечностях:**

**Поведение функции при $ x \to \pm\infty $:**
   - При $ x \to \infty $, $ \frac{1}{x} \to 0 $.
   - Функция $ \sin(t) $ при $ t \to 0 $ стремится к 0, но это не означает, что $ \sin\left(\frac{1}{x}\right) $ имеет предел при $ x \to \infty $, так как $ \frac{1}{x} $ не стремится к 0 монотонно, а принимает все большие и большие значения.
   - Следовательно, $ \sin\left(\frac{1}{x}\right) $ не имеет предела при $ x \to \pm\infty $.

**Следовательно, функция $ f(x) = \sin\left(\frac{1}{x}\right) $ не имеет предела ни в нуле, ни в бесконечностях. Это происходит из-за колебательного характера функции $ \sin(t) $ при $ t \to \infty $ и сложного поведения $ \frac{1}{x} $ при $ x \to 0 $ и $ x \to \infty $.**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def f(x):
    return np.sin(1/x)

x = np.linspace(-1, 1, 200)
x = x[x != 0]

y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции sin(1/x)')
plt.grid(True)
plt.axvline(0, color='black', linewidth=0.5)
plt.axhline(0, color='black', linewidth=0.5)
plt.show()
```

**2. Привести пример функции, не имеющей предела в точке, но определенной в ней.**

Рассмотрим функцию 

$$ f(x) = \begin{cases} 
x + 1 & \text{если } x \geq 0 \\
x - 1 & \text{если } x < 0 
\end{cases} $$

Эта функция не имеет предела в точке $ x = 0 $, так как односторонние пределы $ f(x) $ при $ x \to 0 $ не совпадают. Докажем это формально.

#### Пределы слева и справа от 0

**Предел справа от 0 ($ x \to 0^+ $)**

   Для $ x \geq 0 $, функция $ f(x) $ определяется как $ f(x) = x + 1 $.
   Рассмотрим предел справа:
   $$ \lim_{x \to 0^+} f(x) = \lim_{x \to 0^+} (x + 1). $$
   Поскольку $ x \to 0^+ $, $ x $ стремится к 0, и функция $ x + 1 $ стремится к 1:
   $$ \lim_{x \to 0^+} f(x) = 1. $$

**Предел слева от 0 ($ x \to 0^- $)**

   Для $ x < 0 $, функция $ f(x) $ определяется как $ f(x) = x - 1 $.
   Рассмотрим предел слева:
   $$ \lim_{x \to 0^-} f(x) = \lim_{x \to 0^-} (x - 1). $$
   Поскольку $ x \to 0^- $, $ x $ стремится к 0, и функция $ x - 1 $ стремится к -1:
   $$ \lim_{x \to 0^-} f(x) = -1. $$

#### Вывод

Предел функции $ f(x)$ при $ x \to 0 $ существует, если и только если пределы слева и справа совпадают, т.е.,
$$ \lim_{x \to 0^+} f(x) = \lim_{x \to 0^-} f(x). $$
В нашем случае
$$ \lim_{x \to 0^+} f(x) = 1 \quad \text{и} \quad \lim_{x \to 0^-} f(x) = -1. $$
Поскольку
$$ 1 \neq -1, $$
предел функции  $f(x)$ в точке $x = 0$  не существует.

```python
np.seterr(divide='ignore')

x = np.arange(-10, 10, 0.01)
y = np.where(x >= 0, x + 1, x - 1)

plt.axhline(linewidth=2.0, color="black")
plt.axvline(linewidth=2.0, color="black")
plt.plot(x, y)
plt.plot(0, 1, 'ko', markersize=5)
plt.plot(0, -1, 'wo', markersize=5)
plt.show()
```

**3. Исследовать функцию $f(x) = x^3 - x^2$ по плану:**  
a. Область задания и область значений.  
b. Нули функции и их кратность.  
c. Отрезки знакопостоянства.  
d. Интервалы монотонности.  
e. Четность функции.  
f. Ограниченность.  
g. Периодичность.  

$$ f(x) = x^3 - x^2 $$

**a. Область задания и область значений**
- *Область определения*: Функция $ f(x) = x^3 - x^2 $ определена для всех действительных чисел $ \mathbb{R} $.
- *Область значений*: Область значений функции — все действительные числа $ \mathbb{R} $.

**b. Нули функции и их кратность**

Для $y = 0 $ уравнение $ y = x^3 - x^2 = 0 $ переписываем как:

$$ x^2(x - 1) = 0 $$

 - $ x^2 = 0 $ даёт $ x = 0 $ с кратностью 2, потому что $ x $ появляется дважды в выражении $ x^2 $.  
 - $ x - 1 = 0 $ даёт $ x = 1 $ с кратностью 1, потому что $ x - 1 $ появляется один раз.

**c. Знакопостоянства**  
 Для исследования функции $ y = x^3 - x^2 $ на знакопостоянство, мы должны определить, где функция положительна, отрицательна или равна нулю.

Мы нашли точки, где функция равна нулю: $ x = 0 $ и $ x = 1 $.

Определим знаки функции на интервалах, на которые эти точки делят числовую ось:
   - Интервал $ (-\infty, 0) $:
     Возьмем $ x = -1 $:
     $$ y = (-1)^3 - (-1)^2 = -1 - 1 = -2 $$
     Функция отрицательна на этом интервале.

   - Интервал $ (0, 1) $:
     Возьмем $ x = 0.5 $:
     $$ y = (0.5)^3 - (0.5)^2 = 0.125 - 0.25 = -0.125 $$
     Функция отрицательна на этом интервале.

   - Интервал $ (1, \infty) $:
     Возьмем $ x = 2 $:
     $$ y = 2^3 - 2^2 = 8 - 4 = 4 $$
     Функция положительна на этом интервале.

Сведем результаты в таблицу:
   $
   \begin{array}{c|ccc}
   x & (-\infty, 0) & (0, 1) & (1, \infty) \\
   \hline
   y & - & - & + \\
   \end{array}
   $

Таким образом, функция $ y = x^3 - x^2 $ отрицательна на интервалах $ (-\infty, 0) $ и $ (0, 1) $, и положительна на интервале $ (1, \infty) $.  
В точках $ x = 0 $ и $ x = 1 $ функция равна нулю.

**d. Интервалы монотонности**:
  - На интервале $ (-\infty, 0) $: $ f(x) $ возрастает.
  - На интервале $ (0, \frac{2}{3}) $: $ f(x) $ убывает.
  - На интервале $ (\frac{2}{3}, \infty) $: $ f(x) $ возрастает.

**e. Четность функции**
Функция $ f(x) = x^3 - x^2 $ не является четной и не является нечетной.

**f. Ограниченность**
Функция $ f(x) = x^3 - x^2 $ не ограничена ни сверху, ни снизу.

**g. Периодичность**
Функция $ f(x) = x^3 - x^2 $ не является периодической.


```python
def f(x):
    return x**3 - x**2

np.seterr(divide='ignore')

x = np.arange(-1.5, 1.5, 0.01)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$x^3 - x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции $x^3 - x^2$')
plt.legend()
plt.grid(True)

plt.axvline(x=0, color='black', linewidth=1)
plt.axhline(y=0, color='black', linewidth=1)

plt.xlim(-1.5, 1.5)
plt.ylim(-3, 1)

plt.show()
```

**4. Найти предел:**  
### a. $\lim_{x \to 0} \frac{3x^3 - 2x^2}{4x^2}$<br> ###

Для нахождения предела выполним алгебраические преобразования.
Вынесем за скобки и в числителе и в знаменателе $x^2$:
$$\lim_{x \to 0} \frac{3x^3 - 2x^2}{4x^2} = \lim_{x \to 0} \frac{x^2(3x - 2)}{4x^2}$$

Сократим $x^2$ в числителе и знаменателе. Получим
$$ \lim_{x \to 0} \frac{3x - 2}{4}$$

Подставим $x = 0$ в выражение:
$$\frac{3 \cdot 0 - 2}{4} = \frac{-2}{4} = -\frac{1}{2}$$

Таким образом, предел равен:
$$\lim_{x \to 0} \frac{3x^3 - 2x^2}{4x^2} = -\frac{1}{2}$$

### b.\* $\lim_{x \to 0} \frac{\sqrt{1+x} - 1}{\sqrt[3]{1+x} - 1}$ ###

Числитель $\sqrt{1+x} - 1$ умножим и разделим на сопряженное:

$$\sqrt{1+x} - 1 = \frac{(\sqrt{1+x} - 1)(\sqrt{1+x} + 1)}{\sqrt{1+x} + 1} = \frac{(1+x) - 1}{\sqrt{1+x} + 1} = \frac{x}{\sqrt{1+x} + 1}$$

Преобразуем знаменатель $\sqrt[3]{1+x} - 1$, используя формулу разности кубов:

$$a^3 - b^3 = (a - b)(a^2 + ab + b^2)$$
Здесь $a = \sqrt[3]{1+x}$ и $b = 1$.  
Домножим и числитель и знаменатель на $(a^2 + ab + b^2)$. Таким образом, получим:

$$\sqrt[3]{1+x} - 1 = \frac{(\sqrt[3]{1+x} - 1)((\sqrt[3]{1+x})^2 + \sqrt[3]{1+x} \cdot 1 + 1^2)}{(\sqrt[3]{1+x})^2 + \sqrt[3]{1+x} + 1} = \frac{(1+x) - 1}{(\sqrt[3]{1+x})^2 + \sqrt[3]{1+x} + 1} = \frac{x}{(\sqrt[3]{1+x})^2 + \sqrt[3]{1+x} + 1}$$

Теперь подставим полученные выражения в предел:

$$\lim_{x \to 0} \frac{\sqrt{1+x} - 1}{\sqrt[3]{1+x} - 1} = \lim_{x \to 0} \frac{\frac{x}{\sqrt{1+x} + 1}}{\frac{x}{(\sqrt[3]{1+x})^2 + \sqrt[3]{1+x} + 1}} = \lim_{x \to 0} \frac{x}{\sqrt{1+x} + 1} \cdot \frac{(\sqrt[3]{1+x})^2 + \sqrt[3]{1+x} + 1}{x}$$

$x$ сокращается:

$$= \lim_{x \to 0} \frac{(\sqrt[3]{1+x})^2 + \sqrt[3]{1+x} + 1}{\sqrt{1+x} + 1}$$

Оценим предел при $x \to 0$:

$$= \frac{(\sqrt[3]{1+0})^2 + \sqrt[3]{1+0} + 1}{\sqrt{1+0} + 1} = \frac{1^2 + 1 + 1}{1 + 1} = \frac{3}{2}$$

Таким образом, предел равен:

$$\lim_{x \to 0} \frac{\sqrt{1+x} - 1}{\sqrt[3]{1+x} - 1} = {\frac{3}{2}}$$

### c.\* $\lim\limits_{x \to \infty}\left(\frac{x+3}{x}\right)^{4x+1}$ ###


Введем новую переменную $t = \frac{x+3}{x}$. Тогда $t \to 1$ при $x \to \infty$.
Заметим, что $t = 1 + \frac{3}{x}$. Теперь нам нужно найти предел:
$$\lim\limits_{x \to \infty} t^{4x+1}$$

Запишем $t^{4x+1}$ в виде экспоненты:
$$t^{4x+1} = \exp\left((4x+1) \ln t\right)$$

Найдем предел выражения внутри экспоненты:
$$\lim\limits_{x \to \infty} (4x+1) \ln t$$
Так как $t = 1 + \frac{3}{x}$, то $\ln t \approx \ln\left(1 + \frac{3}{x}\right) \approx \frac{3}{x}$ при $x \to \infty$.

$$\lim\limits_{x \to \infty} (4x+1) \cdot \frac{3}{x} = \lim\limits_{x \to \infty} \left(4 + \frac{1}{x}\right) \cdot 3 = 4 \cdot 3 = 12$$

Возвратимся к экспоненте:
$$\lim\limits_{x \to \infty} \exp\left((4x+1) \ln t\right) = \exp(12) = e^{12}$$

Таким образом, предел равен:
$$\lim\limits_{x \to \infty}\left(\frac{x+3}{x}\right)^{4x+1} = {e^{12}}$$

## ***Тема “Теоремы о пределах”***

### a. $\lim\limits_{x \to 0} \frac{\sin(2x)}{4x}$ ###

Используем первый замечательный предел $\lim\limits_{u \to 0} \frac{\sin(u)}{u} = 1$.  
Пусть $u = 2x$, тогда $u \to 0$ при $x \to 0$.

$$\lim\limits_{x \to 0} \frac{\sin(2x)}{4x} = \lim\limits_{u \to 0} \frac{\sin(u)}{2u} = \frac{1}{2} \lim\limits_{u \to 0} \frac{\sin(u)}{u} = \frac{1}{2} \cdot 1 = \frac{1}{2}$$

Таким образом, предел равен:
$$\lim\limits_{x \to 0} \frac{\sin(2x)}{4x} = \frac{1}{2}$$



### b. $\lim\limits_{x \to 0} \frac{x}{\sin(x)}$ ###

Используем первый замечательный предел $\lim\limits_{u \to 0} \frac{\sin(u)}{u} = 1$.

$$\lim\limits_{x \to 0} \frac{x}{\sin(x)} = \left( \lim\limits_{x \to 0} \frac{\sin(x)}{x} \right)^{-1} = 1^{-1} = 1$$

Таким образом, предел равен:
$$\lim\limits_{x \to 0} \frac{x}{\sin(x)} = 1$$



### c. $\lim\limits_{x \to 0} \frac{x}{\arcsin(x)}$ ###

Используем следствие из первого замечательного предела $\lim\limits_{u \to 0} \frac{\arcsin(u)}{u} = 1$.

$$\lim\limits_{x \to 0} \frac{x}{\arcsin(x)} = \left( \lim\limits_{x \to 0} \frac{\arcsin(x)}{x} \right)^{-1} = 1^{-1} = 1$$

Таким образом, предел равен:
$$\lim\limits_{x \to 0} \frac{x}{\arcsin(x)} = 1$$



### d. $\lim\limits_{x \to \infty} \left(\frac{4x+3}{4x-3}\right)^{6x}$ ###

Введем новую переменную $t = \frac{4x+3}{4x-3}$. Тогда $t \to 1$ при $x \to \infty$.

Заметим, что $t = 1 + \frac{6}{4x-3}$. Теперь нам нужно найти предел:
$$\lim\limits_{x \to \infty} t^{6x}$$

Запишем $t^{6x}$ в виде экспоненты:
$$t^{6x} = \exp\left(6x \ln t\right)$$

Найдем предел выражения внутри экспоненты:
$$\lim\limits_{x \to \infty} 6x \ln t$$
Так как $t = 1 + \frac{6}{4x-3}$, то $\ln t \approx \ln\left(1 + \frac{6}{4x-3}\right) \approx \frac{6}{4x-3}$ при $x \to \infty$.

$$\lim\limits_{x \to \infty} 6x \cdot \frac{6}{4x-3} = \lim\limits_{x \to \infty} \frac{36x}{4x-3} = 9$$

Возвратимся к экспоненте:
$$\lim\limits_{x \to \infty} \exp\left(6x \ln t\right) = \exp(9) = e^{9}$$

Таким образом, предел равен:
$$\lim\limits_{x \to \infty} \left(\frac{4x+3}{4x-3}\right)^{6x} = e^{9}$$



### e.* $\lim\limits_{x \to \infty} \frac{\sin(x) + \ln(x)}{x}$ ###

Используем свойства функций $\sin(x)$ и $\ln(x)$:

$\sin(x)$ ограничена, то есть $|\sin(x)| \leq 1$ для всех $x$.  
$\ln(x)$ растет медленнее, чем $x$, то есть $\frac{\ln(x)}{x} \to 0$ при $x \to \infty$.

Используя эти свойства, мы можем записать:

$$\frac{\sin(x) + \ln(x)}{x} = \frac{\sin(x)}{x} + \frac{\ln(x)}{x}$$

Так как $\frac{\sin(x)}{x} \to 0$ и $\frac{\ln(x)}{x} \to 0$ при $x \to \infty$, то:

$$\lim\limits_{x \to \infty} \frac{\sin(x) + \ln(x)}{x} = 0 + 0 = 0$$

Таким образом, предел равен:

$$\lim\limits_{x \to \infty} \frac{\sin(x) + \ln(x)}{x} = 0$$



### f. * $\lim\limits_{x \to 0} \frac{\sin(x) + \ln(x)}{x}$ ###

Используя замечательные пределы:

1. $\sin(x) \approx x$ при $x \to 0$, поэтому $\frac{\sin(x)}{x} \to 1$.
2. $\ln(x)$ можно представить как $\ln(x) = \ln(1 + (x - 1)) \approx x - 1$ при $x \to 0$, поэтому $\frac{\ln(x)}{x} \approx \frac{x - 1}{x} = 1 - \frac{1}{x}$.

Таким образом, мы можем записать:

$$\frac{\sin(x) + \ln(x)}{x} \approx \frac{x + (x - 1)}{x} = \frac{2x - 1}{x} = 2 - \frac{1}{x}$$

При $x \to 0^+$, $\frac{1}{x} \to \infty$, поэтому:

$$2 - \frac{1}{x} \to -\infty$$

Таким образом предел равен:

$$\lim\limits_{x \to 0} \frac{\sin(x) + \ln(x)}{x} = -\infty$$

```python
def f(x):
    return (np.sin(x) + np.log(x)) / x

x = np.linspace(0.1, 15, 1000)

y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$\frac{\sin(x) + \ln(x)}{x}$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции $\\frac{\\sin(x) + \\ln(x)}{x}$ вблизи $x = 0$')
plt.legend()
plt.grid(True)

plt.axvline(x=0, color='black', linewidth=1)
plt.axhline(y=0, color='black', linewidth=1)

plt.xlim(-1, 15)
plt.ylim(-10, 2)

plt.show()
```
