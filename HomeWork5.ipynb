## Тема 6. Понятие о производной

### 1. Найти производную выражения:

#### a.  $\sin x \cdot \cos x$

Используем правило производной произведения:
$$(\sin x \cdot \cos x)' = (\sin x)' \cdot \cos x + \sin x \cdot (\cos x)'$$
$$= \cos x \cdot \cos x + \sin x \cdot (-\sin x)$$
$$= \cos^2 x - \sin^2 x$$

#### b. $\ln{(2x + 1)^3}$

Используем свойство логарифма и правило дифференцирования сложной функции:
$$\ln{(2x + 1)^3} = 3 \ln{(2x + 1)}$$
$$\frac{d}{dx} [3 \ln{(2x + 1)}] = 3 \cdot \frac{1}{2x + 1} \cdot (2x + 1)'$$
$$= 3 \cdot \frac{1}{2x + 1} \cdot 2$$
$$= \frac{6}{2x + 1}$$

#### c. $\sqrt{\sin^2{(\ln(x^3))}}$

Упростим выражение:
$$\sqrt{\sin^2{(\ln(x^3))}} = |\sin{(\ln(x^3))}|$$

Используем правило дифференцирования сложной функции:
$$\frac{d}{dx} |\sin{(\ln(x^3))}| = \sin{(\ln(x^3))} \cdot \frac{d}{dx} \sin{(\ln(x^3))}$$
$$= \sin{(\ln(x^3))} \cdot \cos{(\ln(x^3))} \cdot \frac{d}{dx} \ln(x^3)$$
$$= \sin{(\ln(x^3))} \cdot \cos{(\ln(x^3))} \cdot \frac{3}{x}$$

Используем формулу двойного угла $\sin(2u) = 2 \sin(u) \cos(u)$:
$$= \frac{1}{2} \sin(2 \ln(x^3)) \cdot \frac{3}{x}$$
$$= \frac{3}{2x} \sin(2 \ln(x^3))$$

#### d. $\frac{x^4}{\ln{(x)}}$

Используем правило дифференцирования частного:
$$\left( \frac{x^4}{\ln x} \right)' = \frac{(x^4)' \cdot \ln x - x^4 \cdot (\ln x)'}{(\ln x)^2}$$
$$= \frac{4x^3 \cdot \ln x - x^4 \cdot \frac{1}{x}}{(\ln x)^2}$$
$$= \frac{4x^3 \ln x - x^3}{(\ln x)^2}$$
$$= \frac{x^3 (4 \ln x - 1)}{(\ln x)^2}$$





c. $$\sqrt{\sin^2{(\ln(x^3))}}= \cos{|\ln{(x^3)|}} \cdot \frac{1}{x^3} \cdot 3x^2 = \frac{3 \cos{|\ln{(x^3)|}}}{x}$$

d. $$\frac{x^4}{\ln{(x)}} = \frac{4x^3\cdot \ln{(x)} - x^4 \frac{1}{x} }{\ln^2{(x)}} = \frac{x^3\ln{(x^4)} - x^3}{\ln^2{(x)}} = \frac{x^3(\ln{(x^4)} - 1)}{\ln^2{(x)}}$$

### 2. Найти выражение производной функции и ее значение в точке:

Для нахождения производной функции $ f(x) = \cos(x^2 + 3x) $ и её значения в точке $ x_0 = \sqrt{\pi} $, мы будем использовать правило дифференцирования сложной функции.

   Функция $ f(x) $ представляет собой композицию двух функций: $ \cos(u) $ и $ u = x^2 + 3x $.

   По правилу дифференцирования сложной функции:
$$f'(x) = \frac{d}{dx} \cos(x^2 + 3x) = -\sin(x^2 + 3x) \cdot \frac{d}{dx}(x^2 + 3x)$$

$$\frac{d}{dx}(x^2 + 3x) = 2x + 3$$

$$f'(x) = -\sin(x^2 + 3x) \cdot (2x + 3)$$

$$f'(\sqrt{\pi}) = -\sin((\sqrt{\pi})^2 + 3\sqrt{\pi}) \cdot (2\sqrt{\pi} + 3)$$

   Упростим выражение внутри синуса:

$$(\sqrt{\pi})^2 + 3\sqrt{\pi} = \pi + 3\sqrt{\pi}$$

   Получаем:

$$f'(\sqrt{\pi}) = -\sin(\pi + 3\sqrt{\pi}) \cdot (2\sqrt{\pi} + 3)$$

   Используем свойство синуса: $ \sin(\pi + x) = -\sin(x) $:

$$\sin(\pi + 3\sqrt{\pi}) = -\sin(3\sqrt{\pi})$$

   Следовательно:

$$f'(\sqrt{\pi}) = -(-(2\sqrt{\pi} + 3)\cdot \sin(3\sqrt{\pi})) $$

$$f'(\sqrt{\pi}) = (2\sqrt{\pi} + 3) \cdot \sin(3\sqrt{\pi})  $$

Оценив
   $$\pi \approx 3.14159$$
   $$\sqrt{\pi} \approx \sqrt{3.14159} \approx 1.772$$
   $$3\sqrt{\pi} \approx 3 \cdot 1.772 \approx 5.316$$

получаем:
   $$f'(\sqrt{\pi}) \approx -5.504$$

### 3. * Найти значение производной функции в точке:


$$ f(x) = \frac{x^3 - x^2 - x - 1}{1 + 2x + 3x^2 - 4x^3} ,  x_0 = 0 $$

Функция $ f(x) $ представляет собой дробь:
$$f(x) = \frac{x^3 - x^2 - x - 1}{1 + 2x + 3x^2 - 4x^3}$$

Для нахождения производной используем правило дифференцирования дроби:
$$f'(x) = \frac{(x^3 - x^2 - x - 1)'(1 + 2x + 3x^2 - 4x^3) - (x^3 - x^2 - x - 1)(1 + 2x + 3x^2 - 4x^3)'}{(1 + 2x + 3x^2 - 4x^3)^2}$$

Найдем производные числителя и знаменателя:
$$(x^3 - x^2 - x - 1)' = 3x^2 - 2x - 1$$
$$(1 + 2x + 3x^2 - 4x^3)' = 2 + 6x - 12x^2$$

Подставим эти производные в формулу:
$$f'(x) = \frac{(3x^2 - 2x - 1)(1 + 2x + 3x^2 - 4x^3) - (x^3 - x^2 - x - 1)(2 + 6x - 12x^2)}{(1 + 2x + 3x^2 - 4x^3)^2}$$

Подставим $ x = 0 $  в выражение для производной:
$$f'(0) = \frac{(3(0)^2 - 2(0) - 1)(1 + 2(0) + 3(0)^2 - 4(0)^3) - (0^3 - 0^2 - 0 - 1)(2 + 6(0) - 12(0)^2)}{(1 + 2(0) + 3(0)^2 - 4(0)^3)^2}$$
$$f'(0) = \frac{(-1)(1) - (-1)(2)}{1^2}$$
$$f'(0) = \frac{-1 + 2}{1}$$
$$f'(0) = 1$$

### 4. Найти угол наклона касательной к графику функции в точке:

$$ f(x) = \sqrt{3x} \ln(x), x_0 = 1 $$

   Функция $ f(x) $ представляет собой произведение двух функций: $ \sqrt{3x} $ и $ \ln(x) $. Используем правило дифференцирования произведения:
   $$ f'(x) = (\sqrt{3x})' \ln(x) + \sqrt{3x} (\ln(x))' $$

   Найдем производные каждой из функций:
   $$ (\sqrt{3x})' = \frac{d}{dx} (3x)^{1/2} = \frac{1}{2} (3x)^{-1/2} \cdot 3 = \frac{3}{2\sqrt{3x}} $$
   $$ (\ln(x))' = \frac{1}{x} $$

   Получаем:
   $$ f'(x) = \frac{3}{2\sqrt{3x}} \ln(x) + \sqrt{3x} \cdot \frac{1}{x} $$

   $$  f'(x) = \frac{3 \ln(x)}{2\sqrt{3x}} + \frac{\sqrt{3x}}{x}  $$
   $$  f'(x) = \frac{3 \ln(x)}{2\sqrt{3x}} + \frac{\sqrt{3}}{\sqrt{x}}  $$

   Подставим $ x = 1 $ в выражение для производной:
   $$  f'(1) = \frac{3 \ln(1)}{2\sqrt{3 \cdot 1}} + \frac{\sqrt{3}}{\sqrt{1}}  $$
   $$  \ln(1) = 0  $$
   $$  f'(1) = \frac{3 \cdot 0}{2\sqrt{3}} + \sqrt{3}  $$
   $$  f'(1) = 0 + \sqrt{3}  $$
   $$  f'(1) = \sqrt{3}  $$

   Угол наклона касательной определяется как арктангенс значения производной в данной точке:
   $$  \alpha = \arctan(\sqrt{3})  = \frac{\pi}{3}  = 60^{\circ}$$
