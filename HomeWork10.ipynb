## Тема 7 “Ряды”

### 1. Исследовать ряд на сходимость, используя признак д’Аламбера:  
$\displaystyle \sum_{n=1}^{\infty} \frac{n^n}{(n!)^2}$ 

Для исследования ряда на сходимость с помощью признака д’Аламбера, мы должны вычислить предел отношения $ \frac{a_{n+1}}{a_n} $, где $ a_n = \frac{n^n}{(n!)^2} $.

$$ a_n = \frac{n^n}{(n!)^2} $$

$$ a_{n+1} = \frac{(n+1)^{n+1}}{((n+1)!)^2} $$

Вычислим отношение
$$ \frac{a_{n+1}}{a_n} = \frac{\frac{(n+1)^{n+1}}{((n+1)!)^2}}{\frac{n^n}{(n!)^2}} = \frac{(n+1)^{n+1}}{((n+1)!)^2} \cdot \frac{(n!)^2}{n^n} $$

$$ \frac{a_{n+1}}{a_n} = \frac{(n+1)^{n+1}}{(n+1)^2 \cdot (n!)^2} \cdot \frac{(n!)^2}{n^n} $$

Сокращаем $ (n!)^2 $ в числителе и знаменателе

$$ \frac{a_{n+1}}{a_n} = \frac{(n+1)^{n+1}}{(n+1)^2} \cdot \frac{1}{n^n} $$

$$ \frac{a_{n+1}}{a_n} = \frac{(n+1)^{n+1}}{(n+1)^2} \cdot \frac{1}{n^n} = \frac{(n+1)^n}{n^n} \cdot \frac{1}{n+1} $$

Получаем

$$ \frac{a_{n+1}}{a_n} = \left( \frac{n+1}{n} \right)^n \cdot \frac{1}{n+1} = \left( 1 + \frac{1}{n} \right)^n \cdot \frac{1}{n+1} $$

Вычислим предел

$$ \lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n \cdot \frac{1}{n+1} $$

Используя второй замечательный предел

$$ \lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n = e ,$$

получаем

$$ \lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n \cdot \frac{1}{n+1} = e \cdot 0 = 0 $$

Поскольку предел отношения $ \frac{a_{n+1}}{a_n} $ равен 0, и 0 < 1, то ряд сходится по признаку д’Аламбера.

#### Ответ: ряд сходится.

### 2. Исследовать ряд на сходимость, используя радикальный признак Коши:  
$\displaystyle \sum_{n=1}^{\infty} \frac{n}{2^n}$  

Для исследования ряда на сходимость с помощью радикального признака Коши, мы должны вычислить предел

$$ \lim_{n \to \infty} \sqrt[n]{\left( \frac{n}{2^n} \right)} $$

Рассмотрим выражение под корнем

$$ \sqrt[n]{\left( \frac{n}{2^n} \right)} = \frac{\sqrt[n]{n}}{2} $$

Теперь нам нужно найти предел $\lim_{n \to \infty} \sqrt[n]{n}$. Понимая, что

$$ \lim_{n \to \infty} \sqrt[n]{n} = 1 ,$$ 

получаем

$$ \lim_{n \to \infty} \frac{\sqrt[n]{n}}{2} = \frac{1}{2} $$

Поскольку $ \frac{1}{2} < 1 $, то согласно радикальному признаку Коши, ряд $ \displaystyle \sum_{n=1}^{\infty} \frac{n}{2^n} $ сходится.

#### Ответ: ряд сходится.

### 3. Исследовать ряд на сходимость, используя признак Лейбница:  
$ \displaystyle \sum_{n=1}^{\infty} \frac{(-1)^n}{n + \ln{n}} $  

Для исследования ряда на сходимость с помощью признака Лейбница, необходимо проверить два условия:

1) Ряд является знакочередующимся.  
2) Члены ряда убывают по модулю: $ \lim_{n \to \infty} |a_n| = 0 $, причём, убывают монотонно.

Последовательность членов ряда по модулю
$$
a_n = \frac{1}{n + \ln{n}}
$$

Нужно показать, что $a_n$ монотонно убывает. Рассмотрим функцию $f(x) = \frac{1}{x + \ln{x}}$ и проверим её на монотонность.

Для этого вычислим производную $f(x)$
$$
f'(x) = \frac{d}{dx} \left( \frac{1}{x + \ln{x}} \right) = -\frac{1 + \frac{1}{x}}{(x + \ln{x})^2} = -\frac{x + 1}{x(x + \ln{x})^2}
$$

Поскольку $x > 0$, $x + 1 > 0$ и $x(x + \ln{x})^2 > 0$, то $f'(x) < 0$. Следовательно, функция $f(x)$ убывает, и последовательность $a_n$ монотонно убывает.

Далее покажем, что
$$
\lim_{n \to \infty} \frac{1}{n + \ln{n}} = 0
$$

Рассмотрим предел
$$
\lim_{n \to \infty} \frac{1}{n + \ln{n}}
$$

Поскольку $n + \ln{n} \to \infty$ при $n \to \infty$, то
$$
\lim_{n \to \infty} \frac{1}{n + \ln{n}} = 0
$$

Таким образом, оба условия признака Лейбница выполнены. Следовательно, ряд $ \displaystyle \sum_{n=1}^{\infty} \frac{(-1)^n}{n + \ln{n}}$ сходится по признаку Лейбница.

#### Ответ: ряд сходится.

### 4. Исследовать ряд на сходимость, используя признак Раабе:  
$ \displaystyle \sum_{n=1}^{\infty} \frac{3^n}{2^n} $  

Для исследования ряда на сходимость с помощью признака Раабе, мы должны вычислить предел:

$$ \lim_{n \to \infty} n \left( \frac{a_n}{a_{n+1}} - 1 \right) $$

$$ a_n = \frac{3^n}{2^n} $$

$$ a_{n+1} = \frac{3^{n+1}}{2^{n+1}} = \frac{3 \cdot 3^n}{2 \cdot 2^n} = \frac{3^{n+1}}{2^{n+1}} $$

$$ \frac{a_n}{a_{n+1}} = \frac{\frac{3^n}{2^n}}{\frac{3^{n+1}}{2^{n+1}}} = \frac{3^n}{2^n} \cdot \frac{2^{n+1}}{3^{n+1}} = \frac{3^n}{2^n} \cdot \frac{2 \cdot 2^n}{3 \cdot 3^n} = \frac{2}{3} $$

Теперь вычислим $ n \left( \frac{a_n}{a_{n+1}} - 1 \right) $

$$ n \left( \frac{2}{3} - 1 \right) = n \left( \frac{2}{3} - \frac{3}{3} \right) = n \left( -\frac{1}{3} \right) = -\frac{n}{3} $$

Найдем предел

$$ \lim_{n \to \infty} n \left( \frac{a_n}{a_{n+1}} - 1 \right) = \lim_{n \to \infty} -\frac{n}{3} = -\infty < 1 $$

Следовательно, ряд расходится по признаку Раабе.

#### Ответ: ряд расходится.

### 6. \* Дана функция: $f(x) = x^2$  

### a. Разложить функцию в ряд Фурье по косинусам на отрезке $x \in [-2; 0]$  
### b. Построить график функции и ее разложения.

a. Период функции $ T = 2 $, так как отрезок имеет длину 2.

Поскольку функция $ y = x^2 $ является четной на заданном интервале, ее разложение в ряд Фурье по косинусам будет иметь вид

$ f(x) = a_0 + \displaystyle \sum_{n=1}^{\infty} a_n \cos\left(\frac{n \pi x}{L}\right) $, где $ L = \frac{T}{2} = 1 $.

Найдем коэффициенты $ a_0 $ и $ a_n $

$$ a_0 = \frac{1}{L} \int_{-2}^{0} x^2 \, dx $$

$$ a_0 = \frac{1}{1} \int_{-2}^{0} x^2 \, dx = \left[ \frac{x^3}{3} \right]_{-2}^{0} = \left( 0 - \left( -\frac{8}{3} \right) \right) = \frac{8}{3} $$

$$ a_n = \frac{2}{L} \int_{-2}^{0} x^2 \cos\left(\frac{n \pi x}{L}\right) \, dx $$

Подставим $ L = 1 $

$$ a_n = 2 \int_{-2}^{0} x^2 \cos(n \pi x) \, dx $$

Для вычисления этого интеграла используем метод интегрирования по частям. Пусть $ u = x^2 $ и $ dv = \cos(n \pi x) \, dx $. Тогда $ du = 2x \, dx $ и $ v = \frac{\sin(n \pi x)}{n \pi} $.

$$ \int u \, dv = uv - \int v \, du $$

$$ \int_{-2}^{0} x^2 \cos(n \pi x) \, dx = \left. x^2 \frac{\sin(n \pi x)}{n \pi} \right|_{-2}^{0} - \int_{-2}^{0} \frac{\sin(n \pi x)}{n \pi} \cdot 2x \, dx $$

Первый член равен нулю, так как $ \sin(0) = 0 $ и $ \sin(-2n \pi) = 0 $

$$ \left. x^2 \frac{\sin(n \pi x)}{n \pi} \right|_{-2}^{0} = 0 $$

Рассмотрим второй интеграл $ - \frac{2}{n \pi} \int_{-2}^{0} x \sin(n \pi x) \, dx $

Снова применим интегрирование по частям. Пусть $ u = x $ и $ dv = \sin(n \pi x) \, dx $. Тогда $ du = dx $ и $ v = -\frac{\cos(n \pi x)}{n \pi} $

$$ \int_{-2}^{0} x \sin(n \pi x) \, dx = \left. -x \frac{\cos(n \pi x)}{n \pi} \right|_{-2}^{0} + \int_{-2}^{0} \frac{\cos(n \pi x)}{n \pi} \, dx $$

Отсюда
$$ \left. -x \frac{\cos(n \pi x)}{n \pi} \right|_{-2}^{0} = -\frac{0 \cdot \cos(0)}{n \pi} + \frac{2 \cdot \cos(-2n \pi)}{n \pi} = \frac{2 \cdot \cos(-2n \pi)}{n \pi} = \frac{2 \cdot (-1)^n}{n \pi} $$

$$ \int_{-2}^{0} \frac{\cos(n \pi x)}{n \pi} \, dx = \frac{1}{(n \pi)^2} \left. \sin(n \pi x) \right|_{-2}^{0} = \frac{1}{(n \pi)^2} \left( \sin(0) - \sin(-2n \pi) \right) = 0 $$

Получаем

$$ a_n = 2 \left( \frac{2 \cdot (-1)^n}{n \pi} \right) = \frac{4 \cdot (-1)^n}{n \pi} $$

Таким образом, разложение функции $ y = x^2 $ в ряд Фурье по косинусам на отрезке $[-2, 0]$ имеет вид

#### Ответ: $ f(x) = \frac{8}{3} + 4 \displaystyle \sum_{n=1}^{\infty} \frac{(-1)^n}{n \pi} \cos(n \pi x) $

b. График функции и ее разложения построим средствами Python.

```python
import numpy as np
import matplotlib.pyplot as plt

def fourier_cosine_series(x, n_terms):
    a0 = 4 * np.pi**2 / 3
    result = a0 / 2
    for n in range(1, n_terms + 1):
        an = 4 * (-1)**n / n**2
        result += an * np.cos(n * x)
    return result

x = np.linspace(-2, 0, 1000)
f_x = x**2
f_fourier = fourier_cosine_series(x, 20)

plt.figure(figsize=(10, 6))
plt.plot(x, f_x, label='$ f(x) = x^2 $', color='red')
plt.plot(x, f_fourier, label='Ряд Фурье по косинусам (20 членов)', linestyle='--', color='blue')
plt.title('Разложение функции $ f(x) = x^2 $ в ряд Фурье по косинусам')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()
plt.grid(True)
plt.show()
```

### 1. Найти неопределенный интеграл:
$$ \int (2x^2 - 2x - 1 + \sin x - \cos x + \ln x + e^x) \, dx $$

Решение:
$$ \int (2x^2 - 2x - 1 + \sin x - \cos x + \ln x + e^x) \, dx = \frac{2x^3}{3} - x^2 - x + \cos x + \sin x + x \ln x - x + e^x + C $$

#### Ответ: $ \frac{2x^3}{3} - x^2 - x + \cos x + \sin x + x \ln x - x + e^x + C $

### 2. Найти неопределенный интеграл:
$$ \int (2x + 6xz^2 - 5x^2y - 3\ln z) \, dx $$

Решение:
$$ \int (2x + 6xz^2 - 5x^2y - 3\ln(z)) \, dx = (6z^2 + 2) \int x \, dx - 5y \int x^2 \, dx - 3\ln z \int 1 \, dx $$
$$ = \frac{(6z^2 + 2)x^2}{2} - \frac{5yx^3}{3} - 3\ln(z)x + C $$

#### Ответ: $ \frac{(6z^2 + 2)x^2}{2} - \frac{5yx^3}{3} - 3\ln(z)x + C $

### 3. Вычислить определенный интеграл:
$$ \int_{0}^{\pi} 3x^2 \sin(2x) \, dx $$

Решение:  
Используем метод интегрирования по частям.
$$ \int u \, dv = uv - \int v \, du $$

$$ u = 3x^2 \quad \text{, } \quad dv = \sin(2x) \, dx \quad \text{, } \quad  du = 6x \, dx \quad \text{, } \quad  v = \int \sin(2x) \, dx = -\frac{1}{2} \cos(2x) $$

Подставляем $ u $, $ du $, $ v $ и $ dv $
$$ \int_{0}^{\pi} 3x^2 \sin(2x) \, dx = \left[ 3x^2 \cdot \left( -\frac{1}{2} \cos(2x) \right) \right]_{0}^{\pi} - \int_{0}^{\pi} \left( -\frac{1}{2} \cos(2x) \right) \cdot 6x \, dx = \left[ -\frac{3x^2}{2} \cos(2x) \right]_{0}^{\pi} + 3 \int_{0}^{\pi} x \cos(2x) \, dx $$

Первое слагаемое
$$
\left[ -\frac{3x^2}{2} \cos(2x) \right]_{0}^{\pi} = -\frac{3\pi^2}{2} \cos(2\pi) - \left( -\frac{3 \cdot 0^2}{2} \cos(0) \right) = -\frac{3\pi^2}{2} \cdot 1 - 0 = -\frac{3\pi^2}{2}
$$

Второе слагаемое (также интегрируя по частям)
$$ 3 \int_{0}^{\pi} x \cos(2x) \, dx = \int_{0}^{\pi} x \cos(2x) \, dx = \left[ x \cdot \frac{1}{2} \sin(2x) \right]_{0}^{\pi} - \int_{0}^{\pi} \frac{1}{2} \sin(2x) \, dx = 0 - \int_{0}^{\pi} \frac{1}{2} \sin(2x) \, dx = 0 - \frac{1}{4} \left[ -\cos(u) \right]_{0}^{2\pi} = 0 - \frac{1}{4} \left( -\cos(2\pi) + \cos(0) \right) = \frac{1}{4} \left( -1 + 1 \right) = 0 $$

Таким образом
$$ \int_{0}^{\pi} 3x^2 \sin(2x) \, dx = -\frac{3\pi^2}{2} + 0 = -\frac{3\pi^2}{2}$$

#### Ответ: $ -\frac{3\pi^2}{2}$

### 4. Найти неопределенный интеграл:
$$ \int \frac{1}{\sqrt{x+1}} \, dx $$

Решение:  
Пусть $ u = x + 1 $
Тогда $ du = dx $

Подставим $ u $ и $ du $
$$ \int \frac{1}{\sqrt{x+1}} \, dx = \int \frac{1}{\sqrt{u}} \, du = \int u^{-\frac{1}{2}} \, du $$

$$ \int u^{-\frac{1}{2}} \, du = \frac{u^{-\frac{1}{2}+1}}{-\frac{1}{2}+1} + C = \frac{u^{\frac{1}{2}}}{\frac{1}{2}} + C = 2u^{\frac{1}{2}} + C $$

Возвращаясь к переменной $ x $, получим

$$ 2u^{\frac{1}{2}} + C = 2(x+1)^{\frac{1}{2}} + C $$

#### Ответ: $ 2\sqrt{x+1} + C $
