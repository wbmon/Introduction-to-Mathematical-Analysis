## Урок 8. Производная функции нескольких переменных. Часть 1

### 1. Найти область определения функции
### $$ z = \sqrt{1 - x^3} + \ln(y^2 - 1) $$ ##

Для нахождения области определения функции $ z = \sqrt{1 - x^3} + \ln(y^2 - 1) $ необходимо учесть два условия:

1. Аргумент квадратного корня должен быть неотрицательным, то есть $ 1 - x^3 \geq 0 $.
2. Аргумент логарифмической функции должен быть положительным, то есть $ y^2 - 1 > 0 $.

Рассмотрим первое условие
$$ 1 - x^3 \geq 0 $$
$$ x^3 \leq 1 $$
$$ x \leq 1 $$

Рассмотрим второе условие
$$ y^2 - 1 > 0 $$
$$ y^2 > 1 $$
$$ y > 1 \quad \text{или} \quad y < -1 $$

Таким образом, область определения функции $ z = \sqrt{1 - x^3} + \ln(y^2 - 1) $ задается следующими неравенствами
$$ x \leq 1 $$
$$ y > 1 \quad \text{или} \quad y < -1 $$

Следовательно, область определения функции $ z $ представляет собой пересечение этих условий
$$ \{ (x, y) \mid x \leq 1 \text{ и } (y > 1 \text{ или } y < -1) \} $$

#### Ответ: $\{ (x, y) \mid x \leq 1 \text{ и } (y > 1 \text{ или } y < -1) \}$

### 2. Найти производные первого порядка функции
$$ z = \left( 1 + \frac{\ln x}{\ln y} \right)^3 $$



Найдем частные производные функции $ z $ по переменным $ x $ и $ y $.


Частная производная по $ x $

$$ \frac{\partial z}{\partial x} = \frac{\partial}{\partial x} \left( 1 + \frac{\ln x}{\ln y} \right)^3 $$

Применим правило дифференцирования сложной функции

$$ \frac{\partial z}{\partial x} = 3 \left( 1 + \frac{\ln x}{\ln y} \right)^2 \cdot \frac{\partial}{\partial x} \left( 1 + \frac{\ln x}{\ln y} \right) $$

Поскольку $ \frac{\partial}{\partial x} (1) = 0 $ и $\frac{\partial}{\partial x} \left( \frac{\ln x}{\ln y} \right) = \frac{1}{x \ln y} $:

$$ \frac{\partial z}{\partial x} = 3 \left( 1 + \frac{\ln x}{\ln y} \right)^2 \cdot \frac{1}{x \ln y} $$

Таким образом,

$$ \frac{\partial z}{\partial x} = \frac{3}{x \ln y} \left( 1 + \frac{\ln x}{\ln y} \right)^2 $$

Частная производная по $ y $:

$$ \frac{\partial z}{\partial y} = \frac{\partial}{\partial y} \left( 1 + \frac{\ln x}{\ln y} \right)^3 $$

Применим правило дифференцирования сложной функции

$$ \frac{\partial z}{\partial y} = 3 \left( 1 + \frac{\ln x}{\ln y} \right)^2 \cdot \frac{\partial}{\partial y} \left( 1 + \frac{\ln x}{\ln y} \right) $$

Поскольку $ \frac{\partial}{\partial y} (1) = 0 $ и $\frac{\partial}{\partial y} \left( \frac{\ln x}{\ln y} \right) = -\frac{\ln x}{y (\ln y)^2} $:

$$ \frac{\partial z}{\partial y} = 3 \left( 1 + \frac{\ln x}{\ln y} \right)^2 \cdot \left( -\frac{\ln x}{y (\ln y)^2} \right) $$

Таким образом,

$$ \frac{\partial z}{\partial y} = -\frac{3 \ln x}{y (\ln y)^2} \left( 1 + \frac{\ln x}{\ln y} \right)^2 $$

#### Ответ:
#### $ \frac{\partial z}{\partial x} = \frac{3}{x \ln y} \left( 1 + \frac{\ln x}{\ln y} \right)^2 $, $ \frac{\partial z}{\partial y} = -\frac{3 \ln x}{y (\ln y)^2} \left( 1 + \frac{\ln x}{\ln y} \right)^2 $

### 3. Найти полный дифференциал функции в точке (1,1)

### $$ z = \sqrt{2xy + \cos{\frac{x}{y}}} $$


Для нахождения полного дифференциала функции $ z $ сначала найдем частные производные функции по $ x $ и $ y $.

Обозначим
$$ u = 2xy + \cos{\frac{x}{y}} $$

Тогда
$$ z = \sqrt{u} $$

Применим правило дифференцирования сложной функции

$$ \frac{\partial z}{\partial x} = \frac{\partial z}{\partial u} \cdot \frac{\partial u}{\partial x} $$

где
$$ \frac{\partial z}{\partial u} = \frac{1}{2\sqrt{u}} $$

Теперь найдем $\frac{\partial u}{\partial x}$

$$ u = 2xy + \cos{\frac{x}{y}} $$

$$ \frac{\partial u}{\partial x} = 2y + \frac{\partial}{\partial x} \left( \cos{\frac{x}{y}} \right) $$

$$ \frac{\partial}{\partial x} \left( \cos{\frac{x}{y}} \right) = -\sin{\frac{x}{y}} \cdot \frac{\partial}{\partial x} \left( \frac{x}{y} \right) = -\sin{\frac{x}{y}} \cdot \frac{1}{y} $$

Таким образом,

$$ \frac{\partial u}{\partial x} = 2y - \frac{\sin{\frac{x}{y}}}{y} $$

Тогда,

$$ \frac{\partial z}{\partial x} = \frac{1}{2\sqrt{2xy + \cos{\frac{x}{y}}}} \left( 2y - \frac{\sin{\frac{x}{y}}}{y} \right) $$


Аналогично поступим для $ y $

$$ \frac{\partial u}{\partial y} = 2x + \frac{\partial}{\partial y} \left( \cos{\frac{x}{y}} \right) $$

$$ \frac{\partial}{\partial y} \left( \cos{\frac{x}{y}} \right) = -\sin{\frac{x}{y}} \cdot \frac{\partial}{\partial y} \left( \frac{x}{y} \right) = -\sin{\frac{x}{y}} \cdot \left( -\frac{x}{y^2} \right) = \sin{\frac{x}{y}} \cdot \frac{x}{y^2} $$

Таким образом,

$$ \frac{\partial u}{\partial y} = 2x + \frac{x \sin{\frac{x}{y}}}{y^2} $$

Тогда,

$$ \frac{\partial z}{\partial y} = \frac{1}{2\sqrt{2xy + \cos{\frac{x}{y}}}} \left( 2x + \frac{x \sin{\frac{x}{y}}}{y^2} \right) $$

Найдем полный дифференциал функции в точке (1,1).

Подставим точку $(x, y) = (1, 1)$ в найденные производные.

Для $\frac{\partial z}{\partial x}$ в точке (1,1)

$$ z = \sqrt{2 \cdot 1 \cdot 1 + \cos{\frac{1}{1}}} = \sqrt{2 + \cos{1}} $$

$$ \frac{\partial z}{\partial x} = \frac{1}{2\sqrt{2 + \cos{1}}} \left( 2 \cdot 1 - \frac{\sin{1}}{1} \right) = \frac{1}{2\sqrt{2 + \cos{1}}} \left( 2 - \sin{1} \right) $$

Для $\frac{\partial z}{\partial y}$ в точке (1,1)

$$ \frac{\partial z}{\partial y} = \frac{1}{2\sqrt{2 + \cos{1}}} \left( 2 \cdot 1 + \frac{1 \cdot \sin{1}}{1^2} \right) = \frac{1}{2\sqrt{2 + \cos{1}}} \left( 2 + \sin{1} \right) $$

Теперь полный дифференциал $ dz $ запишем как

$$ dz = \frac{\partial z}{\partial x} dx + \frac{\partial z}{\partial y} dy $$

Подставим найденные частные производные

$$ dz = \frac{1}{2\sqrt{2 + \cos{1}}} (2 - \sin{1}) dx + \frac{1}{2\sqrt{2 + \cos{1}}} (2 + \sin{1}) dy $$

Полный дифференциал функции $ z $ в точке $(1, 1)$ равен

$$ dz = \frac{2 - \sin{1}}{2\sqrt{2 + \cos{1}}} dx + \frac{2 + \sin{1}}{2\sqrt{2 + \cos{1}}} dy $$

#### Ответ: $ dz = \frac{2 - \sin{1}}{2\sqrt{2 + \cos{1}}} dx + \frac{2 + \sin{1}}{2\sqrt{2 + \cos{1}}} dy $

## 4. Исследовать на экстремум функцию:
## $$ z = x^2 + xy + y^2 - 6x - 9y $$

Найдем частные производные первого порядка

$$ \frac{\partial z}{\partial x} = 2x + y - 6 $$

$$ \frac{\partial z}{\partial y} = x + 2y - 9 $$

Приравняем частные производные к нулю и решим систему уравнений

$$ 2x + y - 6 = 0 $$
$$ x + 2y - 9 = 0 $$

Используем метод подстановки. Из первого уравнения
$$ y = 6 - 2x $$

Подставим $ y = 6 - 2x $ во второе уравнение

$$ x + 2(6 - 2x) - 9 = 0 $$
$$ x + 12 - 4x - 9 = 0 $$
$$ -3x + 3 = 0 $$
$$ x = 1 $$

Подставим $ x = 1 $ в первое уравнение $ y = 6 - 2x $

$$ y = 6 - 2 \cdot 1 = 4 $$

Таким образом, критическая точка

$$ (x, y) = (1, 4) $$

Найдем частные производные второго порядка:
$$ \frac{\partial^2 z}{\partial x^2} = 2 $$
$$ \frac{\partial^2 z}{\partial y^2} = 2 $$
$$ \frac{\partial^2 z}{\partial x \partial y} = 1$$

Используем достаточные условия для экстремума:  

Для функции двух переменных $ z = f(x, y) $ в критической точке $(x_0, y_0)$ достаточные условия экстремума 

- Если $\frac{\partial^2 z}{\partial x^2}(x_0, y_0) > 0$ и $\left(\frac{\partial^2 z}{\partial x^2}(x_0, y_0)\right) \left(\frac{\partial^2 z}{\partial y^2}(x_0, y_0)\right) - \left(\frac{\partial^2 z}{\partial x \partial y}(x_0, y_0)\right)^2 > 0$, то функция имеет локальный минимум в точке $(x_0, y_0)$.

- Если $\frac{\partial^2 z}{\partial x^2}(x_0, y_0) < 0$ и $\left(\frac{\partial^2 z}{\partial x^2}(x_0, y_0)\right) \left(\frac{\partial^2 z}{\partial y^2}(x_0, y_0)\right) - \left(\frac{\partial^2 z}{\partial x \partial y}(x_0, y_0)\right)^2 > 0$, то функция имеет локальный максимум в точке $(x_0, y_0)$.

- Если $\left(\frac{\partial^2 z}{\partial x^2}(x_0, y_0)\right) \left(\frac{\partial^2 z}{\partial y^2}(x_0, y_0)\right) - \left(\frac{\partial^2 z}{\partial x \partial y}(x_0, y_0)\right)^2 < 0$, то функция не имеет экстремума в точке $(x_0, y_0)$.

Проверим условия в критической точке $(1, 4)$:

$$ \frac{\partial^2 z}{\partial x^2}(1, 4) = 2 > 0 $$

$$ \left(\frac{\partial^2 z}{\partial x^2}(1, 4)\right) \left(\frac{\partial^2 z}{\partial y^2}(1, 4)\right) - \left(\frac{\partial^2 z}{\partial x \partial y}(1, 4)\right)^2 = 2 \cdot 2 - 1^2 = 4 - 1 = 3 > 0 $$

Так как оба условия выполнены, функция $ z = x^2 + xy + y^2 - 6x - 9y $ имеет локальный минимум в точке $(1, 4)$.

Вычислим значение функции в точке минимума:
$$ z(1, 4) = 1^2 + 1 \cdot 4 + 4^2 - 6 \cdot 1 - 9 \cdot 4 = 1 + 4 + 16 - 6 - 36 = -21 $$

Таким образом, функция $ z = x^2 + xy + y^2 - 6x - 9y $ имеет минимум в точке $(1, 4)$ со значением $-21$.

#### Ответ: $(1, 4)$ — точка минимума, $ z(1, 4) = -21 $.
