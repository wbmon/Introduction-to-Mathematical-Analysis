## Урок 11. Интеграл. Ряды. Часть 2

### 1. Решить уравнения: 


### 1: $ y' - y = 2x - 3 $

Для решения используем метод интегрирующего множителя. Интегрирующий множитель $\mu(x)$ определяется как
$$ \mu(x) = e^{\int p(x) \, dx} = e^{\int -1 \, dx} = e^{-x} $$

Умножаем обе части уравнения на $\mu(x)$
$$ e^{-x} y' - e^{-x} y = (2x - 3)e^{-x} $$

Получаем
$$ \frac{d}{dx} \left( y e^{-x} \right) = (2x - 3)e^{-x} $$

Интегрируем обе части уравнения
$$ y e^{-x} = \int (2x - 3)e^{-x} \, dx $$

$$ u = 2x - 3 \quad \text{и} \quad dv = e^{-x} \, dx $$
$$ du = 2 \, dx \quad \text{и} \quad v = -e^{-x} $$

$$ \int u \, dv = uv - \int v \, du $$
$$ \int (2x - 3)e^{-x} \, dx = (2x - 3)(-e^{-x}) - \int (-e^{-x}) \cdot 2 \, dx $$
$$ = -(2x - 3)e^{-x} + 2 \int e^{-x} \, dx $$
$$ = -(2x - 3)e^{-x} - 2e^{-x} $$
$$ = -e^{-x}(2x - 3 + 2) $$
$$ = -e^{-x}(2x - 1) $$

Таким образом
$$ y e^{-x} = -e^{-x}(2x - 1) + C $$

Умножаем обе части на $ e^x $
$$ y = - (2x - 1) + Ce^x $$
$$ y = -2x + 1 + Ce^x $$

### 2: $ x^2 y' + xy + 1 = 0 $

Преобразуем уравнение
   $$ y' + \frac{y}{x} = -\frac{1}{x^2} $$

$\mu(x) = e^{\int \frac{1}{x} \, dx} = e^{\ln x} = x$.

$$ x y' + y = -\frac{1}{x} $$
$$ \frac{d}{dx} (x y) = -\frac{1}{x} $$
$$ x y = \int -\frac{1}{x} \, dx = -\ln x + C $$

$$ y = \frac{-\ln x + C}{x} $$


#### Ответ: 1. $ y = -2x + 1 + Ce^x $; 2. $ y = \frac{-\ln x + C}{x} $


### 2. Исследовать ряд на сходимость:

### 1: $ \displaystyle \sum_{n=1}^{\infty} \frac{n+2}{n^2+n+1} $

Сравним с гармоническим рядом $ \displaystyle \sum_{n=1}^{\infty} \frac{1}{n}$, который расходится.
   $$  \frac{n+2}{n^2+n+1} \approx \frac{n}{n^2} = \frac{1}{n} \quad \text{при} \quad n \to \infty  $$
Так как $\frac{n+2}{n^2+n+1} \sim \frac{1}{n}$, а ряд $\sum_{n=1}^{\infty} \frac{1}{n}$ расходится, то и исходный ряд расходится.

### 2: $ \displaystyle \sum_{n=1}^{\infty} \frac{n^n}{n!}$

Применяем признак Даламбера
$$ \lim_{n \to \infty} \frac{a_{n+1}}{a_n} = \lim_{n \to \infty} \frac{(n+1)^{n+1}}{(n+1)!} \cdot \frac{n!}{n^n} = \lim_{n \to \infty} \frac{(n+1)^{n+1}}{n^n (n+1)} =$$
  
$$  = \lim_{n \to \infty} \left( \frac{n+1}{n} \right)^n = \lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n = e  $$

Так как предел равен $e > 1$,  ряд расходится.

### 3: $ \displaystyle \sum_{n=1}^{\infty} n \cdot \left(1 - \frac{1}{n}\right)^{n^2} $

Применим предельный признак для исследования сходимости ряда. Рассмотрим предел
$$ \lim_{n \to \infty} n \cdot \left(1 - \frac{1}{n}\right)^{n^2} $$

Используя эквивалентность $ \left(1 - \frac{1}{n}\right)^n \approx \frac{1}{e} $ при $ n \to \infty $:
$$ \left(1 - \frac{1}{n}\right)^{n^2} \approx \left(\frac{1}{e}\right)^n $$

Получаем, что
$$ \lim_{n \to \infty} n \cdot \left(\frac{1}{e}\right)^n = \lim_{n \to \infty} \frac{n}{e^n} = 0 $$

Поскольку предел равен 0, ряд сходится.


### 4: $ \displaystyle \sum_{n=1}^{\infty} \frac{(-1)^{n+1}}{2n - \ln n}$

Используем признак Лейбница для знакопеременных рядов. Проверим два условия:
1. Члены ряда убывают по абсолютной величине: $\left| \frac{1}{2n - \ln n} \right| $
2. Предел членов ряда равен нулю
$$\lim_{n \to \infty} \frac{1}{2n - \ln n} = 0 $$

Т.к. оба условия выполнены, ряд сходится.

### 5: $ \displaystyle \sum_{n=1}^{\infty} \frac{(x-2)^{n+1}}{3^n(n+2)}$

Применяем признак Даламбера
$$  \lim_{n \to \infty} \frac{a_{n+1}}{a_n} = \lim_{n \to \infty} \frac{(x-2)^{n+2}}{3^{n+1}(n+3)} \cdot \frac{3^n(n+2)}{(x-2)^{n+1}} = \lim_{n \to \infty} \frac{(x-2) \cdot (n+2)}{3(n+3)} = \frac{x-2}{3} $$
   - Если $\left|\frac{x-2}{3}\right| < 1$, то ряд сходится.
   - Если $\left|\frac{x-2}{3}\right| > 1$, то ряд расходится.
   - Если $\left|\frac{x-2}{3}\right| = 1$, то необходимо дополнительное исследование.

Таким образом, ряд сходится при $\left|\frac{x-2}{3}\right| < 1$, т.е. при $|x-2| < 3$.
